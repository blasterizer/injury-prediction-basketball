##Creating the feature of fat percentage from BMI ##

```python
##Azura Stevens, Julie Allemand, Satou Sabally didn't have any matches as they did not have any games
##Lou Katie Samuelson : Katie Lou Samuelson;
## 'Diamond Deshields': Diamond DeShields	;
## 'Didi Richards', : DiDi Richards

name_mapping = {
    'Lou Katie Samuelson': 'Katie Lou Samuelson',
    'Diamond Deshields': 'Diamond DeShields',
    'Didi Richards': 'DiDi Richards'
    }

def convert_name(name):
    if name in name_mapping:
        return name_mapping[name]
    else:
        return name
wnba_inj_clean['Athlete'] = wnba_inj_clean['Athlete'].apply(convert_name)
```

```python
set(wnba_inj_clean['Athlete']) - set(wnba_stats['Player'])
```

```python
##Need to remove these 3 rows
wnba_inj_clean = wnba_inj_clean[~wnba_inj_clean['Athlete'].isin(['Azura Stevens', 'Julie Allemand', 'Satou Sabally'])]
set(wnba_inj_clean['Athlete']) - set(wnba_stats['Player'])
```

```python
##Now need to merge the wnba_inj_clean and wnba_stats based on player name
wnba_inj_merged = pd.merge(wnba_inj_clean,wnba_stats,left_on='Athlete',right_on='Player',how='left')
wnba_inj_merged
```

```python
wnba_inj_merged.drop(columns=['Position','Player','Team_y','G.1','GS','MP.1','FG','FG%','3P','3PA','3P%','2P','2PA','2P%','FT','FTA','FT%','ORB','AST','STL','BLK','TOV','PF','PTS'],inplace=True)
wnba_inj_merged
```

```python
set(wnba_inj_merged['Pos'])
```

```python
#Converting the positions
position_conversion = {
    'C': 'C',
    'C-F': 'PF',
    'F': 'SF',
    'F-C': 'PF',
    'F-G': 'SF',
    'G': 'PG',
    'G-F': 'SG'
}
wnba_inj_merged['Pos'] = wnba_inj_merged['Pos'].map(position_conversion)
set(wnba_inj_merged['Pos'])
```

```python
##The minutes played are total minutes played, so need to change to average minutes played and find FGA per minutes played
wnba_inj_merged['MP']= wnba_inj_merged['MP']/wnba_inj_merged['G']
wnba_inj_merged['FGA_per_MP']= wnba_inj_merged['FGA']/wnba_inj_merged['MP']
wnba_inj_merged
```

##In order to keep uniformity with the NBA dataset we use the FGA_per_minutes_played. And here it was total FGA unlike the male dataset, so dividing twice

```python
#rounding the wnba[mp] to 2 places
wnba_inj_merged['MP'] = wnba_inj_merged['MP'].round(2)
wnba_inj_merged['FGA_per_MP'] = wnba_inj_merged['FGA_per_MP']/wnba_inj_merged['MP'].round(2)
wnba_inj_merged
```

```python
corr = wnba_inj_merged[['AGE','body_fat_perct','MP','FGA_per_MP','TRB']].corr()
corr.style.background_gradient(cmap='coolwarm')
```

```python
#Total Rebounds also need to be per minutes played to avoid multicollinearity
wnba_inj_merged['TRB_per_MP'] = wnba_inj_merged['TRB']/wnba_inj_merged['MP']
wnba_inj_merged
```

```python
corr = wnba_inj_merged[['AGE','body_fat_perct','MP','FGA_per_MP','TRB_per_MP']].corr()
corr.style.background_gradient(cmap='coolwarm')
```
